# Proceso de ETL

## 1.- Importacion de Librerias 

```python
# Importar Librerias
import pandas as pd
```

## 2.- Lectura de csv 

```python
# Lectuar de todos los csv's
df_03_Local = pd.read_csv('Tasas de Interes/01_03_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_04_Local = pd.read_csv('Tasas de Interes/03_04_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_05_Local = pd.read_csv('Tasas de Interes/01_05_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_06_Local = pd.read_csv('Tasas de Interes/01_06_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_07_Local = pd.read_csv('Tasas de Interes/03_07_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_08_Local = pd.read_csv('Tasas de Interes/01_08_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_09_Local = pd.read_csv('Tasas de Interes/01_09_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
df_10_Local = pd.read_csv('Tasas de Interes/02_10_2023_MonedaLocal.csv', encoding = 'latin-1', sep=";")
listaBancos = pd.read_csv('Tasas de Interes\ListaBancos.csv',encoding='utf-8',sep=";")
```

## 3.- Union de cv's

```python
# Unions de los csv, sin tomar en cuenta las cabeceras de los siguientes archivos
dfGeneralBancos = pd.concat([df_03_Local,
                     df_04_Local,
                     df_05_Local,
                     df_06_Local,
                     df_07_Local,
                     df_08_Local,
                     df_09_Local,
                     df_10_Local], ignore_index=True)
```

```python
# Vista general del dataframe Unido
dfGeneralBancos
```

## 4.- Transformaciones 

### 4.1 Transformaciones a columnas

```python
# Seleccionar las columnas sin conversion a float y luego convetir las que si necesitan
columnasEstaticas = ['Tasa Anual (%)', 'Fecha','Tipo de Empresa']
for columnas in dfGeneralBancos.columns:
    if columnas not in columnasEstaticas:
        dfGeneralBancos[columnas] =pd.to_numeric(dfGeneralBancos[columnas],errors = 'coerce')

# Convertir las fechas a formato fecha %d/%m/%Y
dfGeneralBancos['Fecha'] = pd.to_datetime(dfGeneralBancos['Fecha'], format='%d/%m/%Y')['Fecha'] = pd.to_datetime(dfGeneralBancos['Fecha'], format='%d/%m/%Y')

# Eliminar columnas que no siven
dfGeneralBancos = dfGeneralBancos.drop(['Unnamed: 21'],axis= 1)
```

### 4.2 Filtros de los datos a usar, tipo de empresa y plazo de prestamo 

```python
# Seleccion valor por tipo de empresa 
tipoDeEmpresa = 'Medianas Empresas'
dfGeneralBancos = dfGeneralBancos[dfGeneralBancos['Tipo de Empresa'] ==tipoDeEmpresa ]
# Filtros a usar por tipo de prestamo
# Se elimino los espacios en blanco porque no se encontraban los datos buscados
dfGeneralBancos['Tasa Anual (%)'] = dfGeneralBancos['Tasa Anual (%)'].str.replace(r'\s', '', regex=True)
tipoDePrestamo = 'Préstamosamásde360días'
dfGeneralBancos= dfGeneralBancos[dfGeneralBancos['Tasa Anual (%)'] == tipoDePrestamo]
dfGeneralBancos
```

### 4.3 Hallar los promedios mensuales y convertirlos a data frame

```python
# Sacar promedios solo a las columnas que tienen como data Floats (solo las tasas)
promedio_por_columna = dfGeneralBancos.select_dtypes(include=['float']).mean(skipna=True)
# Convertir la lista de promedios a dataframe 
df_promedios = pd.DataFrame({'Promedio de bancos a 360 dias': promedio_por_columna})
# Ordenarlos de menor a mayor
df_promedios = df_promedios.sort_values(by = 'Promedio de bancos a 360 dias',ascending= True)
```

```python
df_promedios
```

### 4.4 Hallar la tasas de los bancos de las fechas mas proximas 

```python
# Hallar la fecha mas reciente
fechaReciente = dfGeneralBancos['Fecha'].max()
# Filtro de fecha mas proxima
dfUltimaFecha = dfGeneralBancos[dfGeneralBancos['Fecha']== fechaReciente]
# Eliminar las columnas que no interesan
columnasElim = ['Tasa Anual (%)','Fecha','Tipo de Empresa']
dfUltimaFecha = dfUltimaFecha.drop(columnasElim, axis = 1)
# Invertir las columnas para obtener los bancos en filas 
dfUltimaFecha =dfUltimaFecha.transpose()
# Convetir la fecha a str para concatenarla al nombre del cabecera 
fechaRecienteStr = fechaReciente.strftime('%d/%m/%Y')
dfUltimaFecha = dfUltimaFecha.rename(columns= {305:"Tasa a 360 dias "+ fechaRecienteStr })
```

```python
dfUltimaFecha
```

### 4.5 Insetar indices de las tablas como columnas 

```python
df_promedios.reset_index(level=0,inplace=True)
dfUltimaFecha.reset_index(level=0,inplace=True)
```

### 4.6 Unir las columnas promedios de fechas y tasas mas recientes

```python
dfAnalizado = pd.merge(dfUltimaFecha, df_promedios, on="index")
# Renombrar
dfAnalizado.columns = dfAnalizado.columns.str.replace("index", "Bancos para Empresa Mediana")
dfAnalizado 
```

### 4.7.- Traer los datos que al menos tengan un no nulo, o la ultima fecha o el promedio 

```python
num_nulos_por_fila = dfAnalizado.isnull().sum(axis=1)
# Elimina las filas que contienen más de un nulo
df_filtradoAnalizado = dfAnalizado[num_nulos_por_fila <= 1]
df_filtradoAnalizado =df_filtradoAnalizado.sort_values(by = 'Tasa a 360 dias 02/10/2023',ascending= True)
df_filtradoAnalizado
```

### 4.8.- Exportar datos para jefe inmediato 

```python
df_filtradoAnalizado.to_csv('df_filtradoAnalizado.csv', index=False)
```
